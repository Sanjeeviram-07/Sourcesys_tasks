***Create a DataFrame***

```python
import pandas as pd
data = {
    "Name": ["Arun", "Kiran", "Meena"],
    "Age": [22, 21, 23],
    "City": ["Chennai", "Madurai", "Coimbatore"]
}
df = pd.DataFrame(data)
print(df)
```

***Display First Rows of DataFrame***

```python

data = {
    "Name": ["Ravi", "Sita", "Anil", "Deepa"],
    "Age": [20, 22, 21, 23]
}

df = pd.DataFrame(data)

print(df.head())
```

***Display Last Rows***

```python

data = {
    "Product": ["Pen", "Book", "Pencil", "Bag"],
    "Price": [10, 50, 5, 300]
}

df = pd.DataFrame(data)

print(df.tail())
```

***Select a Column***

```python

data = {
    "Name": ["Amit", "Rahul", "Priya"],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)

print(df["Marks"])
```

***Add a New Column***

```python

data = {
    "Name": ["Ravi", "Sita", "Anil"],
    "Marks": [80, 85, 90]
}

df = pd.DataFrame(data)

df["Grade"] = ["B", "B", "A"]

print(df)
```

***Filter Data in DataFrame***

```python

data = {
    "Name": ["Kumar", "Arun", "Rani", "Vijay"],
    "Age": [21, 24, 20, 23]
}

df = pd.DataFrame(data)

print(df[df["Age"] > 21])
```

***Find Mean Value***

```python

data = {
    "Marks": [70, 80, 90, 85]
}

df = pd.DataFrame(data)

print("Average:", df["Marks"].mean())
```

***Sorting DataFrame***

```python

data = {
    "Name": ["A", "B", "C"],
    "Score": [75, 90, 60]
}

df = pd.DataFrame(data)

print(df.sort_values("Score"))
```
