***STRING PROGRAMS***

```python
text = "data science"

print("Upper:", text.upper())
print("Lower:", text.lower())
```

```python
print("Length:", len(text))
print("Count of 'a':", text.count('a'))
```

```python
print("Replaced:", text.replace("data", "AI"))
```

***LIST PROGRAMS***

```python
nums = [5, 6, 7]
nums.append(8)
nums.insert(2, 20)
print("Updated List:", nums)
```

```python
nums = [12, 4, 9, 2]
nums.sort()
print("Sorted:", nums)
nums.reverse()
print("Reversed:", nums)
```

```python
nums = [50, 60, 70, 80]
nums.pop()
print("After Pop:", nums)
```

***PATTERN PROGRAMS***

```python
n = 6
for i in range(n):
    for j in range(n):
        print("#", end=" ")
    print()
```

```python
n = 4
for i in range(1, n+1):
    for j in range(i):
        print("#", end=" ")
    print()
```

```python
n = 10
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```
