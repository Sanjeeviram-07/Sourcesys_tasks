***Create DataFrame Using NumPy Array***

```python
import numpy as np
import pandas as pd
data = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,90]])

df = pd.DataFrame(data, columns=["A","B","C"])

print(df)
```

***Apply NumPy Sum on DataFrame***

```python
data = {"Marks":[60,70,80,90]}

df = pd.DataFrame(data)

total = np.sum(df["Marks"])

print("Total Marks:", total)
```

***Find Mean Using NumPy***

```python
data = {"Scores":[50,65,75,85]}

df = pd.DataFrame(data)

mean_value = np.mean(df["Scores"])

print("Mean Score:", mean_value)
```

***Apply Square Operation Using NumPy***

```python
data = {"Numbers":[1,2,3,4]}

df = pd.DataFrame(data)

df["Square"] = np.square(df["Numbers"])

print(df)
```

***Find Maximum Value***

```python
data = {"Values":[15,45,10,60,30]}

df = pd.DataFrame(data)

max_val = np.max(df["Values"])

print("Maximum:", max_val)
```

***Find Standard Deviation***

```python
data = {"Marks":[55,65,75,85,95]}

df = pd.DataFrame(data)

std_val = np.std(df["Marks"])

print("Standard Deviation:", std_val)
```

***Apply NumPy Log Function***

```python
data = {"Numbers":[1,2,3,4,5]}

df = pd.DataFrame(data)

df["Log"] = np.log(df["Numbers"])

print(df)
```

***Element-wise Addition Using NumPy***

```python
data = {"A":[1,2,3],
        "B":[4,5,6]}

df = pd.DataFrame(data)

df["Sum"] = np.add(df["A"], df["B"])

print(df)
```

***Multiply Column Values***

```python
data = {"Price":[100,200,300],
        "Quantity":[2,3,4]}

df = pd.DataFrame(data)

df["Total"] = np.multiply(df["Price"], df["Quantity"])

print(df)
```

***Convert DataFrame to NumPy Array***

```python
data = {"X":[10,20,30],
        "Y":[40,50,60]}

df = pd.DataFrame(data)

arr = df.to_numpy()

print(arr)
```
