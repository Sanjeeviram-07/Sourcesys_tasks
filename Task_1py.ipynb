## ***ARITHMETIC PROGRAMS***

```python
a = 20
b = 6
print("Modulus:", a % b)
print("Floor Division:", a // b)
```

```python
num = 5
print("Square:", num ** 2)
print("Cube:", num ** 3)
```

***LIST PROGRAMS***

```python
nums = [12, 5, 30, 7]
print("Smallest:", min(nums))
```

```python
nums = [1,2,3,4,5]
nums.reverse()
print("Reversed List:", nums)
```

```python
nums = [1,2,3,4,5,6]
evens = [x for x in nums if x % 2 == 0]
print("Even Numbers:", evens)
```

***DICTIONARY PROGRAMS***

```python
car = {"brand":"Toyota","year":2022}
print(car.get("brand"))
```

```python
car = {"brand":"Toyota","year":2022}
car["year"] = 2024
print(car)
```

```python
student = {"name":"Siva","age":21,"course":"AI"}
print("Total Keys:", len(student))
```

***CONDITIONAL PROGRAMS***

```python
num = 7
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
age = 18
if age >= 18:
    print("Eligible to Vote")
else:
    print("Not Eligible")
```

```python
temp = 35
if temp > 30:
    print("Hot")
else:
    print("Normal Weather")
```

***LOOP PROGRAMS***

```python
for i in range(2,11,2):
    print(i)
```

```python
total = 0
for i in range(1,6):
    total += i

print("Total:", total)
```

```python
n = 5
while n > 0:
    print(n)
    n -= 1
```

***OOP PROGRAMS***

```python
class Student:
    def __init__(self,name):
        self.name = name

    def display(self):
        print("Student Name:", self.name)

s = Student("sanju")
s.display()
```

```python
class Vehicle:
    def start(self):
        print("Vehicle Started")

class Car(Vehicle):
    def drive(self):
        print("Car is Driving")

c = Car()
c.start()
c.drive()
```

```python
class Laptop:
    def __init__(self,price):
        self.__price = price

    def get_price(self):
        return self.__price

l = Laptop(50000)
print("Price:", l.get_price())
```
