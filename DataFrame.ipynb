```python
data = [
    {"Name": "Arun", "Age": 22, "City": "Chennai"},
    {"Name": "Meena", "Age": 21, "City": "Madurai"},
    {"Name": "Ravi", "Age": 23, "City": "Salem"}
]

for row in data:
    print(row)
```

```python
data = [
    {"Name": "Arun", "Age": 22},
    {"Name": "Meena", "Age": 21},
    {"Name": "Ravi", "Age": 23}
]

ages = [row["Age"] for row in data]

print("Ages:", ages)
```

```python
data = [
    {"Name": "Arun", "Marks": 80},
    {"Name": "Meena", "Marks": 90},
    {"Name": "Ravi", "Marks": 85}
]

for row in data:
    row["Grade"] = "Pass"

print(data)
```

```python
data = [
    {"Name": "Arun", "Age": 22},
    {"Name": "Meena", "Age": 19},
    {"Name": "Ravi", "Age": 24}
]

result = [row for row in data if row["Age"] > 20]

print(result)
```

```python
data = [
    {"Name": "A", "Marks": 70},
    {"Name": "B", "Marks": 80},
    {"Name": "C", "Marks": 90}
]

total = 0
for row in data:
    total += row["Marks"]

avg = total / len(data)

print("Average:", avg)
```

```python
data = [
    {"Name": "Arun", "Age": 22},
    {"Name": "Meena", "Age": 19},
    {"Name": "Ravi", "Age": 24}
]

data.sort(key=lambda x: x["Age"])

print(data)
```
